```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Path to your shapefile
# us_eco_l3.shp

shp_file_path = './rawDatasets/EcoRegions/us_eco_l3.shp'

# Read the shapefile
gdf = gpd.read_file(shp_file_path)

# Display all columns in the DataFrame
print(gdf.columns)
print(gdf.count)

print("Top 5 Records:")
print(gdf.head(10))

```

```python
# Plot the data
gdf.plot()

# Show the plot
plt.show()
```

```python
# Export to GeoJSON
geojson_path = './processedDatasets/EcoRegions/EcoRegions.geojson'
gdf.to_file(geojson_path, driver='GeoJSON')
```
