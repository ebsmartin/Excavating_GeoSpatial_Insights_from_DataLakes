```python
import os
import geopandas as gpd
import matplotlib.pyplot as plt

# Directory containing your Shapefiles
shp_directory = './rawDatasets/GeologicalFeatures/GMNA_SHAPES'

# Loop through all files in the directory
for filename in os.listdir(shp_directory):
    if filename.endswith('.shp'):
        shp_file_path = os.path.join(shp_directory, filename)
        gdf = gpd.read_file(shp_file_path)

        # Convert and save to GeoJSON
        geojson_path = './processedDatasets/GeologicalFeatures/' + filename.replace('.shp', '.geojson')
        gdf.to_file(geojson_path, driver='GeoJSON')
        print(f'Converted {filename} to GeoJSON')

        # Plotting the GeoDataFrame
        gdf.plot()
        plt.title(filename)
        plt.show()

        # Print the dimensions of the GeoDataFrame
        print(f'Dimensions of {filename}: {gdf.shape}')
```
