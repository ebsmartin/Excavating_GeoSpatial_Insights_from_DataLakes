    atlas_stco: Full Agricultural Census state-county equivalent code. This code uniquely identifies a county or a county-equivalent region. The code is derived from standards set by the American National Standards Institute (ANSI).

    atlas_name: Lowercase name of the county equivalent (or agriculture area). This is the name of the county or the agricultural area as defined in the 2017 Census of Agriculture.

    atlas_caps: Uppercase name of the county equivalent (or agriculture area). Similar to atlas_name, but in uppercase for consistency or display purposes.

    atlas_area: Land area in square miles of a county equivalent (or agriculture area). This represents the total land area of the county or the agricultural census area.

    atlas_acre: Land area in acres of a county equivalent (or agriculture area). This is another measure of the area, similar to atlas_area, but in acres.

    entity: Type of county-level entity. This could be a county, parish, municipio, or an Agricultural Census Area, especially relevant in places like Alaska, Hawaii, and Puerto Rico.

    cntya: Agricultural Census county equivalent character codes. These are character codes that represent counties or county-equivalents in the Agricultural Census.

    cntyn: Agricultural Census county equivalent numeric codes. This is the numeric representation of the county or county-equivalent regions as used in the Agricultural Census.

    geometry: Feature geometry. This column stores the geometric data (like polygons) that represent the shape and location of each county or region in the dataset.

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Path to your shapefile
# CoUS17_WGS84WMAS.shp

shp_file_path = '../rawDatasets/AgriculturalArea/CoUS17_WGS84WMAS.shp'

# Read the shapefile
gdf = gpd.read_file(shp_file_path)
```

```python
import pandas as pd

county_names = '../rawDatasets/AgriculturalArea/CountyNames.csv'
variable_lookup = '../rawDatasets/AgriculturalArea/VariableLookup.csv'
crops_and_plants = '../rawDatasets/AgriculturalArea/CropsAndPlants.csv'
economics = '../rawDatasets/AgriculturalArea/Economics.csv'
farms = '../rawDatasets/AgriculturalArea/Farms.csv'
livestock_and_animals = '../rawDatasets/AgriculturalArea/LivestockAndAnimals.csv'
producers = '../rawDatasets/AgriculturalArea/Producers.csv'

county_names_df = pd.read_csv(county_names)
variable_lookup_df = pd.read_csv(variable_lookup)
crops_and_plants_df = pd.read_csv(crops_and_plants)
economics_df = pd.read_csv(economics)
farms_df = pd.read_csv(farms)
livestock_and_animals_df = pd.read_csv(livestock_and_animals)
producers_df = pd.read_csv(producers)

print(county_names_df.head())
print(variable_lookup_df.head())
print(crops_and_plants_df.head())
print(economics_df.head())
print(farms_df.head())
print(livestock_and_animals_df.head())
print(producers_df.head())
```

```python
# Convert atlas_stco in the geodataframe to string
gdf['atlas_stco'] = gdf['atlas_stco'].astype(str)

# Also, ensure that FIPSTEXT in each CSV dataframe is of type string
county_names_df['FIPSTEXT'] = county_names_df['FIPSTEXT'].astype(str).str.pad(width=5, side='left', fillchar='0')
crops_and_plants_df['FIPSTEXT'] = crops_and_plants_df['FIPSTEXT'].astype(str).str.pad(width=5, side='left', fillchar='0')
economics_df['FIPSTEXT'] = economics_df['FIPSTEXT'].astype(str).str.pad(width=5, side='left', fillchar='0')
farms_df['FIPSTEXT'] = farms_df['FIPSTEXT'].astype(str).str.pad(width=5, side='left', fillchar='0')
livestock_and_animals_df['FIPSTEXT'] = livestock_and_animals_df['FIPSTEXT'].astype(str).str.pad(width=5, side='left', fillchar='0')
producers_df['FIPSTEXT'] = producers_df['FIPSTEXT'].astype(str).str.pad(width=5, side='left', fillchar='0')
```

```python
gdf = gdf.merge(county_names_df, left_on='atlas_stco', right_on='FIPSTEXT', how='left')
```

```python
gdf = gdf.merge(crops_and_plants_df, on='FIPSTEXT', how='left', suffixes=('', '_cp'))
```

```python
gdf = gdf.merge(economics_df, on='FIPSTEXT', how='left', suffixes=('', '_econ'))
```

```python
gdf = gdf.merge(farms_df, on='FIPSTEXT', how='left', suffixes=('', '_farms'))
```

```python
gdf = gdf.merge(livestock_and_animals_df, on='FIPSTEXT', how='left', suffixes=('', '_la'))
```

```python
gdf = gdf.merge(producers_df, on='FIPSTEXT', how='left', suffixes=('', '_producers'))
```
